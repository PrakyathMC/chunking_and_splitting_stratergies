```python
from langchain.text_splitter import CharacterTextSplitter

text = "LangChain makes it easy to work with large documents by splitting them into smaller chunks."

# Split into chunks of 30 characters with 10 character overlap
splitter = CharacterTextSplitter(separator = "",chunk_size=30, chunk_overlap=10)
chunks = splitter.split_text(text)

print(chunks)
```
